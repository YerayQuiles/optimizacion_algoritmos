##DevoluciÃ³n de cambio por tÃ©cnica voraz




```
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION

cambio_monedas(15,[25,10,5,1])
```

## N-Reinas por tÃ©cnica de vueta atrÃ¡s


```
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")

def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la soluciÃ³n tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " estÃ¡ " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(8)
```

```
def n_reinas(n):
    """Resuelve el problema de N-Reinas usando backtracking"""
    soluciones = []

    def es_valida(pos, col):
        for i in range(col):
            if pos[i] == pos[col] or abs(pos[i] - pos[col]) == col - i:
                return False
        return True

    def resolver(pos, col):
        if col == n:
            soluciones.append(pos[:])
            return

        for fila in range(n):
            pos[col] = fila
            if es_valida(pos, col):
                resolver(pos, col + 1)

    resolver([0] * n, 0)
    return soluciones

def mostrar(solucion):
    """Muestra un tablero con la soluciÃ³n"""
    n = len(solucion)
    for i in range(n):
        print(' '.join('â™›' if solucion[j] == i else 'Â·' for j in range(n)))
    print()

# Ejecutar para 8 reinas
sols = n_reinas(8)
print(f"Total de soluciones para 8 reinas: {len(sols)}\n")

# Mostrar las primeras 5 soluciones
for i, sol in enumerate(sols[:5]):
    print(f"SoluciÃ³n {i+1}: {[x+1 for x in sol]}")
    mostrar(sol)
```

```
import numpy as np

TARIFAS = [
    [0, 5, 4, 3, 999, 999, 999],
    [999, 0, 999, 2, 3, 999, 11],
    [999, 999, 0, 1, 999, 4, 10],
    [999, 999, 999, 0, 5, 6, 9],
    [999, 999, 999, 999, 0, 999, 4],
    [999, 999, 999, 999, 999, 0, 3],
    [999, 999, 999, 999, 999, 999, 0]
]

def viaje_optimizado(tarifas):
    """
    VersiÃ³n optimizada usando Floyd-Warshall modificado.
    Complejidad: O(nÂ³) pero con mejor constante y uso de memoria.
    """
    n = len(tarifas)
    INF = 999

    # Usar numpy para operaciones mÃ¡s rÃ¡pidas
    dist = np.array(tarifas, dtype=np.int32)
    next_node = np.arange(n).reshape(1, -1).repeat(n, axis=0)

    # Floyd-Warshall optimizado para grafos dirigidos acÃ­clicos
    for k in range(n):
        # VectorizaciÃ³n parcial para mejor rendimiento
        for i in range(k):  # Solo necesitamos i < k por ser DAG
            if dist[i][k] < INF:
                # OperaciÃ³n vectorizada en j
                mask = (dist[i][k] + dist[k]) < dist[i]
                dist[i][mask] = dist[i][k] + dist[k][mask]
                next_node[i][mask] = next_node[k][mask]

    return dist, next_node

def reconstruir_ruta_optimizada(next_node, origen, destino):
    """ReconstrucciÃ³n de ruta optimizada sin lista intermedia."""
    if next_node[origen][destino] == destino:
        return [origen, destino]

    ruta = [origen]
    actual = origen
    while actual != destino:
        actual = next_node[actual][destino]
        ruta.append(actual)
    return ruta

# VersiÃ³n sin numpy para mÃ¡xima eficiencia en Python puro
def viaje_rapido(tarifas):
    """VersiÃ³n mÃ¡s rÃ¡pida sin dependencias externas."""
    n = len(tarifas)
    INF = 999

    # Trabajar directamente sobre la matriz original cuando sea posible
    dist = [row[:] for row in tarifas]  # Copia superficial mÃ¡s rÃ¡pida
    next_node = list(range(n)) * n
    next_node = [next_node[i*n:(i+1)*n] for i in range(n)]

    # OptimizaciÃ³n: aprovechar que es un DAG (grafo dirigido acÃ­clico)
    for k in range(1, n):
        for i in range(k):
            if dist[i][k] < INF:
                # Desenrollar el bucle interno para mejor rendimiento
                for j in range(k + 1, n):
                    nuevo_costo = dist[i][k] + dist[k][j]
                    if nuevo_costo < dist[i][j]:
                        dist[i][j] = nuevo_costo
                        next_node[i][j] = next_node[k][j]

    return dist, next_node

# Ejecutar versiÃ³n optimizada
import time

# Benchmark
start = time.time()
dist, next_node = viaje_rapido(TARIFAS)
tiempo_rapido = time.time() - start

print("ðŸš€ RESULTADOS OPTIMIZADOS\n")
print(f"Tiempo de ejecuciÃ³n: {tiempo_rapido*1000:.3f} ms\n")

# Mostrar algunos resultados
print("COSTOS MÃNIMOS (primeras 3 filas):")
for i in range(3):
    print(f"Desde {i}: {[d if d < 999 else 'âˆž' for d in dist[i]]}")

print("\nRUTAS Ã“PTIMAS:")
for origen, destino in [(0, 6), (1, 5), (2, 4)]:
    ruta = reconstruir_ruta_optimizada(next_node, origen, destino)
    print(f"{origen} â†’ {destino}: {' â†’ '.join(map(str, ruta))} (Costo: {dist[origen][destino]})")

# FunciÃ³n para consultas rÃ¡pidas
def costo_minimo(origen, destino):
    return dist[origen][destino] if dist[origen][destino] < 999 else None

print(f"\nCosto mÃ­nimo 0â†’6: {costo_minimo(0, 6)}")
```
